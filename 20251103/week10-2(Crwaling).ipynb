```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
req = requests.get("https://dhlottery.co.kr/store.do?method=topStore&pageGubun=L645&wiselog=H_D_2_1")
```

```python
html = req.text
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
name = soup.select("#article > div:nth-child(2) > div > div:nth-child(4) > table > tbody > tr:nth-child(1) > td:nth-child(2)")
```

```python
name[0].text
```

```python
name = soup.select("#article > div:nth-child(2) > div > div:nth-child(4) > table > tbody > tr > td:nth-child(2)")
```

```python
name_t = []
for n in name:
    # print(n.text)
    name_t.append(n.text)
```

```python
name_t
```

```python
addr = soup.select("#article > div:nth-child(2) > div > div:nth-child(4) > table > tbody > tr > td:nth-child(4)")
```

```python
addr_t = []
for n in addr:
    addr_t.append(n.text)
```

```python
addr_t
```

```python
gubun = soup.select("#article > div:nth-child(2) > div > div:nth-child(4) > table > tbody > tr > td:nth-child(3)")
```

```python
gubun_t = []
for n in gubun: 
    gubun_t.append(n.text.strip())
```

```python
gubun_t
```

```python
print(name_t[0], '|', gubun_t[0], '|',  addr_t[0])
```

```python
df = pd.DataFrame(list(zip(name_t, gubun_t, addr_t)), columns=['상호명', '구분', '주소'])
```

```python
df.to_csv('lotto1103.csv', index=False, encoding = 'utf-8-sig')
```
